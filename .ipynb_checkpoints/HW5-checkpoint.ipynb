```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.patches as patches

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')
```

```python
# Read CSV
mouse_data = "mouse_drug_data.csv"
mouse_data_df = pd.read_csv(mouse_data)
clinical_data_df =pd.read_csv("clinicaltrial_data.csv")
# Merge two data frames together and display the combined data table for preview
combined_data = pd.merge(mouse_data_df, clinical_data_df, on="Mouse ID")
combined_data.head()
```

```python
# check missing data
clean_combined_data= combined_data.dropna(how="any")
clean_combined_data.count()
```

```python
# Part 1 Tumor Response to Treatment
# # Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
data_group = combined_data.groupby(["Drug","Timepoint"])
tumor_volume_mean = data_group["Tumor Volume (mm3)"].mean()
# Convert it to DataFrame
tumor_volume_mean_df = pd.DataFrame(tumor_volume_mean).reset_index()
tumor_volume_mean_df.head()

# my result only show drug name in the first row, the HW sample showed drug name in everyrow,
# how can i make drug name in everyrow? by adding reset_index()
```

```python
#calculate standard deviation
tumor_volume_sd = data_group["Tumor Volume (mm3)"].sem()
tumor_volume_sd_df = pd.DataFrame(tumor_volume_sd).reset_index()
tumor_volume_sd_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_volume_mean_df_reshape = pd.pivot_table(tumor_volume_mean_df, values=["Tumor Volume (mm3)"], index=["Timepoint"], columns=["Drug"])
# Preview that Reformatting worked

tumor_volume_mean_df_reshape.head()
```

```python
type(tumor_volume_mean_df_reshape)
tumor_volume_mean_df_reshape.columns
```

```python
# Part 1 Tumor Response to Treatment
# # Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

tumor_volume_mean_df_reshape1 = pd.pivot_table(tumor_volume_mean_df, values=["Tumor Volume (mm3)"], index=["Timepoint"], columns=["Drug"])["Tumor Volume (mm3)"]
# Preview that Reformatting worked

tumor_volume_mean_df_reshape1
```

```python
sd_df = pd.pivot_table(tumor_volume_sd_df, values=["Tumor Volume (mm3)"], index=["Timepoint"], columns=["Drug"])["Tumor Volume (mm3)"]
sd_df
```

```python
x = tumor_volume_mean_df_reshape1.reset_index()['Timepoint']
x
```

```python
plt.errorbar(x, tumor_volume_mean_df_reshape1["Capomulin"], yerr = sd_df["Capomulin"],  marker='o', markersize=8, color='r', ecolor = 'r',capsize = 5, capthick = 2)
plt.errorbar(x, tumor_volume_mean_df_reshape1["Infubinol"], yerr = sd_df["Infubinol"],  marker='^', markersize=8, color='g', ecolor = 'g',capsize = 5, capthick = 2)
plt.errorbar(x, tumor_volume_mean_df_reshape1["Ketapril"], yerr = sd_df["Ketapril"],  marker='x', markersize=8, color='b', ecolor = 'b',capsize = 5, capthick = 2)
plt.errorbar(x, tumor_volume_mean_df_reshape1["Placebo"], yerr = sd_df["Placebo"],  marker='s', markersize=8, color='y', ecolor = 'y',capsize = 5, capthick = 2)
plt.title("Tumor Response to Treatment")
plt.legend(loc='best')
plt.xlabel("Time(Days)")
plt.ylabel("Tumor Volume(mm3)")
plt.savefig('1.png'),
plt.show()
```

```python
# Part 2 Metastatic Response to Treatment
```

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
# Convert to DataFrame
# Preview DataFrametumor_volume_mean_df = pd.DataFrame(tumor_volume_mean)
metastatic_sites_mean = data_group["Metastatic Sites"].mean()
metastatic_sites_mean_df = pd.DataFrame(metastatic_sites_mean)
metastatic_sites_mean_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastatic_sites_mean_df_reshape = pd.pivot_table(metastatic_sites_mean_df, values=["Metastatic Sites"], index=["Timepoint"], columns=["Drug"])["Metastatic Sites"]

# Preview that Reformatting worked
metastatic_sites_mean_df_reshape.head()
```

```python
x_m = metastatic_sites_mean_df_reshape.reset_index()['Timepoint']
x_m
```

```python
#calculate standard deviation
metastatic_sites_sd = data_group["Metastatic Sites"].sem()
metastatic_sites_sd_df = pd.DataFrame(metastatic_sites_sd).reset_index()

# Minor Data Munging to Re-Format the Data Frames
metastatic_sites_sd_df_reshape = pd.pivot_table(metastatic_sites_sd_df, values=["Metastatic Sites"], index=["Timepoint"], columns=["Drug"])["Metastatic Sites"]

# Preview that Reformatting worked
metastatic_sites_sd_df_reshape.head()
```

```python
# Generate the Plot (with Error Bars)

# Save the Figure

# Show the Figure
plt.errorbar(x_m, metastatic_sites_mean_df_reshape["Capomulin"], yerr = metastatic_sites_sd_df_reshape["Capomulin"],  marker='o', markersize=8, color='r', ecolor = 'r',capsize = 5, capthick = 2)
plt.errorbar(x_m, metastatic_sites_mean_df_reshape["Infubinol"], yerr = metastatic_sites_sd_df_reshape["Infubinol"],  marker='^', markersize=8, color='g', ecolor = 'g',capsize = 5, capthick = 2)
plt.errorbar(x_m, metastatic_sites_mean_df_reshape["Ketapril"], yerr = metastatic_sites_sd_df_reshape["Ketapril"],  marker='x', markersize=8, color='b', ecolor = 'b',capsize = 5, capthick = 2)
plt.errorbar(x_m, metastatic_sites_mean_df_reshape["Placebo"], yerr = metastatic_sites_sd_df_reshape["Placebo"],  marker='s', markersize=8, color='y', ecolor = 'y',capsize = 5, capthick = 2)
plt.title("Metastatic Spread During Treatment")
plt.legend(loc='best')
plt.xlabel("Time(Days)")
plt.ylabel("Metastatic Sites")
plt.savefig('2.png'),
plt.show()
```

```python
# Part 3 Survival Rates
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_number = data_group["Mouse ID"].count()
mouse_number.head()
# Convert to DataFrame
mouse_number_df = pd.DataFrame(mouse_number)
mouse_number_df.head()
# Preview DataFrame
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_number_df_reshape = pd.pivot_table(mouse_number_df, values=["Mouse ID"], index=["Timepoint"], columns=["Drug"])["Mouse ID"]
# Preview the Data Frame
mouse_number_df_reshape.head()
```

```python
x_s =mouse_number_df_reshape.reset_index()['Timepoint']
x_s
```

```python
# Generate the Plot (Accounting for percentages)
# Save the Figure
# Show the Figure
plt.errorbar(x_s, mouse_number_df_reshape["Capomulin"], marker='o', markersize=8, color='r', ecolor = 'r',capsize = 5, capthick = 2,)
plt.errorbar(x_s, mouse_number_df_reshape["Infubinol"], marker='^', markersize=8, color='g', ecolor = 'g',capsize = 5, capthick = 2)
plt.errorbar(x_s, mouse_number_df_reshape["Ketapril"], marker='x', markersize=8, color='b', ecolor = 'b',capsize = 5, capthick = 2)
plt.errorbar(x_s, mouse_number_df_reshape["Placebo"],  marker='s', markersize=8, color='y', ecolor = 'y',capsize = 5, capthick = 2)
plt.title("Mouse Survival During Treatment")
plt.legend(loc='best')
plt.xlabel("Time(Days)")
plt.ylabel("Survival Rate(%)")
plt.savefig('3.png'),
plt.show()
```

```python
#Summary Bar Graph
# Calculate the percent changes for each drug and display the data to confirm
the_percent_changes = (tumor_volume_mean_df_reshape1.pct_change(periods = 9))*100
the_percent_changes
```

```python
# Store all Relevant Percent Changes into a Tuple
the_percent_changes_df = pd.DataFrame(the_percent_changes.loc[45,:]).reset_index()
the_percent_changes_df
```

```python
# Splice the data between passing and failing drugs
four_drug_percent_change = the_percent_changes_df.loc[[0,2,3,5],["Drug", 45]]
four_drug_percent_change

```

```python
x_p = np.arange(len(four_drug_percent_change["Drug"]))
x_p
```

```python
y_p = four_drug_percent_change[45]
y_p.reset_index()
```

```python
four_drug_percent_change["color_p"] = ['green' if x <0 else 'red' for x in four_drug_percent_change[45]]
four_drug_percent_change["color_p"].reset_index()
```

```python
plot = plt.bar(x_p, y_p, color=four_drug_percent_change["color_p"], alpha=0.75, align="edge",width=-1, linewidth=20)
plt.grid()
tick_locations = [value for value in x_p]
plt.xticks(tick_locations, ["Capomulin", "Infubinol", "Ketapril", "Placebo"])

# Sets the x limits of the current chart
plt.xlim(-1, len(x_p)-0.25)
# Sets the y limits of the current chart
plt.ylim(-20, max(y_p)+10)
plt.title("Tumor Change Over 45 Days")
plt.xlabel("Drug")
plt.ylabel("% Tumor Volume Changes")

# Annotate Text
for value in plot:
    height = value.get_height()
    plt.text(value.get_x() + value.get_width()/2.,
             1.002*height, '%d' % int(height), ha='center', va='bottom', color = 'black')
plt.show()  
plt.tight_layout()
```
